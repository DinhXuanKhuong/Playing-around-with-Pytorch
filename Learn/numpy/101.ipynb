```python
petallength = iris[:, 2].astype(float)
small = (petallength < 3)
medium = (petallength >= 3) & (petallength < 5)
large = (petallength >= 5)

petallength = petallength.astype(str)
petallength[small] = "small"
petallength[medium] = "medum"
petallength[large] = "large"

petallength
```

```python
petallength = iris[:, 2].astype(float)

petallength_binned = np.digitize(petallength, [0, 3, 5, np.inf])

map_name = {1: "small", 2: "medium", 3: "large", 4: np.nan}

res = [map_name[x] for x in petallength_binned]
res
```
