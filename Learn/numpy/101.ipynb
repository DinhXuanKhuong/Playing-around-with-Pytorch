```python
iris_2d = np.array([row.tolist()[:4]for row in iris_1d]).astype(float)
iris_2d[:5]
```

```python
mean_sepallength = np.mean(iris_2d[:, 0])
median_sepallength = np.median(iris_2d[:, 0])
std_sepallength = np.std(iris_2d[:, 0])

print(mean_sepallength, median_sepallength, std_sepallength)
```

```python
sepallength = iris_2d[:, 0].copy()
np.set_printoptions(precision=3)
norm_sepallength = (sepallength - sepallength.min()) / (sepallength.max() - sepallength.min() + 1e-12)
norm_sepallength[:5]
```
